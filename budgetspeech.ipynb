Importing Required Libraries

```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from PIL import Image 
```

extracting the words from the text document and storing it in array

```python
arr = []
with open('budget_speech.txt','r',encoding="utf8") as file:
   
    # reading each line    
    for line in file:
   
        # reading each word        
        for word in line.split():
           # print(word)
   
            # displaying the words           
            arr.append(word)
```

delimiting the words with semi-colon

```python
text = ";".join(arr)
```

creating the mask for visulaization by converting the image into array

```python
mask = np.array(Image.open('india.jpg'))
```

creating a function that converts text into black

```python
def black_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    return("hsl(0,100%, 1%)")
```

creating wordcloud using wordcloud library and using stopwords to skip certain words 

```python
wordcloud = WordCloud(background_color="white", width=3000, height=2000, mask=mask,collocations=False,
                      stopwords=['has','been''such','has been','will','per cent','and','of','or','in','i','I','the','have','from','it','also','for','be','a','onby','on','with','This','is','to','are','our','by','an','as','all','that','at','through','we','per','cent','their','which','these','other','year','up','provide','being','years','has been']).generate(text)
```

using the black_color_func to change the font color

```python
wordcloud.recolor(color_func = black_color_func)
```

setting the size of the figure and displaying it

```python
plt.figure(figsize=[15,10])
plt.axis("off")
plt.imshow(wordcloud)
plt.plot()
```
